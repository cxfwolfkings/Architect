**Series**

- 一维带标签的数组
- 创建：

  - pd.Series({"":,"":})
- 索引

**DataFrame**

- 二维的series容器
- 创建
  - pd.DataFrame(np.arange(12).reshape((3,4)),index=,columns=)
  - pd.DataFrame([{},{},{}])
  - pd.DataFrame("anem":[],"age":[])
  - 如果字典中有的键不存在，该位置为nan
- 索引和切片
  - 简单的索引切片
    - df[1]：dataframe中写数字，取行
    - df[3:]
    - df["Y"]：写字符串，取列
  - loc
    - 按照便签进行索引
    - df.loc[["a","c"],["d"]]
  - iloc
    - 按照位置进行索引
    - df.iloc[[2,3],3]
- bool索引
  - df[df["a"]>100]
  - 多条件情况
    - df[()|()]：或
    - df[()&()]：且
- 缺失数据的填充
  - 0的缺失
    - 判断是否为缺失值
    - df[df==0]=np.nan
  - nan的缺失
    - pd.isnull()
    - pd.notnull()
    - 填充
      - df["a"]=df["a"].fillna(df["a"].mean()) // 填充具体a这一列
      - df.fillna(df.mean())
    - 删除行
      - df.dropna(axis=0,how="any[all]",inplace="True")

**join**

- 按照index来进行分组
- t1.join(t2)

**merge**

- 按照列数据进行分组
- t1.merge(t2,on="a",how='inner')  // 内连接
- t1.merge(t2,on="a",how="outter")  // 并集
- t1.merge(t2,on="a",how='left')  // 以t1为准
- t1.merge(t2,on="a",how='right')  // 以t2为准
- t1.merge(t2,left_on="a",right_on="b",how='right')

**分组聚合**

- df.groupby(by="")：得到groupby对象，能够调用聚合方法，能够遍历
- df.groupby(by).mean()
- df.groupby(by).count()
- df.groupby(by=["",""]).count()：带复合索引的DataFrame

**索引**

- df.index
- df.index=["a","c"]  // 设置索引的值
- df.set_index("a")  // 设置某一列为索引
- df.set_index(["a","c"])
- 从复合索引中取值
  - Series
    - s1["a"]["b"]
    - s1["a","b"]
  - DataFrame
    - df.loc["a"].loc["b"]
- 从里层索引开始选择
  - df.swaplevel()
  - 交换里外层索引的位置
